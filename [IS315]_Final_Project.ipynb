# RDF

```python
import matplotlib.pyplot as plt
```

```python
rdf_files = [
    ('rdf_DIPC_POP2.xvg', 'DIPC-POP2'),
    ('rdf_CHOL_POP2.xvg', 'CHOL-POP2'),
    ('rdf_DPPC_POP2.xvg', 'DPPC-POP2')
]

for filename, label in rdf_files:
    r, g = [], []
    with open(filename) as f:
        for line in f:
            if line.startswith(('#', '@')):
                continue
            parts = line.split()
            r.append(float(parts[0]))
            g.append(float(parts[1]))
    plt.plot(r, g, label=label)

plt.xlabel('r (nm)')
plt.ylabel('g(r)')
plt.title('Radial Distribution Functions')
plt.legend()
plt.show()
```

# MSD

```python
msd_files = [
    ('msd_CHOL.xvg', 'CHOL'),
    ('msd_DIPC.xvg', 'DIPC'),
    ('msd_DPPC.xvg', 'DPPC'),
    ('msd_PIP2.xvg', 'PIP2')
]

for filename, label in msd_files:
  ts, msd = [], []
  with open(filename) as f:
    for line in f:
      if line.startswith(('#', '@')):
        continue

      parts = line.split()
      ts.append(float(parts[0]))
      msd.append(float(parts[1]))


  plt.plot(ts, msd, label=label)

plt.xlabel('t (ps)')
plt.ylabel('MSD (nm^2)')
plt.title('Mean Square Displacement')
plt.legend()
```
